```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

import pyrtools as pt
%load_ext autoreload
%autoreload 2
```

```python
img_tmp=plt.imread('./Einstein.jpg')
img_curie=plt.imread('./Curie.pgm')
r,g,b=[img_tmp[:,:,i] for i in range(3)]
im=(r*0.299+g*0.587+b*0.114).astype(np.uint8)
```

```python
binom5=pt.binomial_filter(5)
lo_filt=2*binom5*binom5.T
blurred1=pt.corrDn(image=im,filt=lo_filt,step=(2,2))
fine1=im-pt.upConv(image=blurred1,filt=lo_filt,step=(2,2))
blurred2=pt.corrDn(image=img_curie,filt=lo_filt,step=(2,2))
blurred2=pt.upConv(image=blurred2,filt=lo_filt,step=(2,2))
pt.imshow([fine1,blurred1],title=['fine1','blurred1'])
hybrid=fine1+blurred2
pt.imshow(hybrid)
```
